# **Mapping the Margins: An In-depth Analysis of Out-of-School Youth**

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('dataset.csv', encoding='latin-1')


stat = data['Status']

data.plot(x='Countries', y=['Male', 'Female'], kind="bar", figsize=(20, 18))
plt.title("Male vs. Female Out-of-School Youth")


plt.savefig('gender.png')
plt.xlabel("Country")
plt.ylabel("Number of Male and Female")
```

```python
data.plot(x='Countries', y='Total', kind="bar", figsize=(17, 15), color="purple")
plt.title("Out-of-School Youth Worldwide")
plt.savefig('worldwide.png')
plt.xlabel("Country")
plt.ylabel("Total of Out-of-School Youth")
```

```python
data.info()
```

```python
reg = data['Region']
pd.Series(reg).value_counts().sort_index(ascending=True).plot(kind='bar', color='magenta')
plt.savefig('region.png')
plt.xlabel("Region")
plt.ylabel("Number of Country with Out-of-School Youth Record")
```

```python
pd.Series(stat).value_counts().sort_index(ascending=True).plot(kind='bar', color='pink')
plt.title("Status of Countries with Out-of-School Youth")
plt.savefig('status.png')
plt.xlabel("Status")
plt.ylabel("Number of Country with Out-of-School Youth Record")
```

```python
yr = data['Year']
pd.Series(yr).value_counts() .sort_index(ascending=True).plot(kind='bar', color='green')
plt.title("Highest Yearly Record")
plt.savefig('yearly.png')
plt.xlabel("Year")
plt.ylabel("Number of Country with Highest Out-of-School Youth Record")
```

# **Immediate and Direct Findings**


*   Number of female out-of-school youth has outnumbered the number of males. **This suggest that female out-of-school youth is more than male out-of-school youth worldwide.**
*   SSA region garnered greatest number of countries with the record of out-of-school youth. **This suggests that SSA region has the most number of out-of-school youth.**

*   Least developed and less developed countries outnumbered the number of countries with the record of out-of-school youth that belongs to most developed countries.  **This suggests that most number of out-of-school youth came from the less developed countries.**
*   Most countries has the highest record of out-of-school youth in the year 2018. **This suggests that the highest number of out-of-school youth worldwide is recorded in 2018.**
